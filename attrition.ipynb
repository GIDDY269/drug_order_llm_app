```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Create a figure with a specific size
plt.figure(figsize=[20, 10])

for i, col in enumerate(df.columns):
    plt.subplot(8, 3, i+1)
    plt.hist(df[col])
    
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
