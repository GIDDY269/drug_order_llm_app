```python
def get_desc(section_url,img_link):
    '''
    A function to get the descriptions of the products
    '''
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    
    driver = Chrome(options=chrome_options)
    driver.get('https://glovoapp.com' + section_url)
    page_source = driver.page_source

    cookie = WebDriverWait(driver,20).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@class='onetrust-close-btn-handler onetrust-close-btn-ui banner-close-button ot-close-icon']"))
        ).click()



    print(img_link['src'])
    folder = WebDriverWait(driver,20).until(
        EC.element_to_be_clickable((By.XPATH, f"//img[@src='{img_link['src']}']"))
    )
    folder.click()

    path_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, 'path[data-v-00c725c8]'))
    ).click()

    driver.implicitly_wait(5)
    desc = driver.find_element(By.XPATH,"//span[@class='product-details__description--cut-text']").text

    des_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, 'path[data-v-00c725c8]'))
    ).click()

    driver.quit()

    return desc
```
